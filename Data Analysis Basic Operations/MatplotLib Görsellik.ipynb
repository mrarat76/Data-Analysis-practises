```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1=np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2= numpyDizisi1 ** 2
```

```python
(benimFigur, benimEksen) = plt.subplots()
benimEksen.plot(numpyDizisi1, numpyDizisi2,"g",alpha=0.3) ##alpha saydamlık derecesidir azaldıkça saydamlık artar.
benimEksen.plot(numpyDizisi2, numpyDizisi1, color="#3A95A8") ## renk kodu)
```

```python
(yeniFigur, yeniEksen) = plt.subplots()
yeniEksen.plot (numpyDizisi1, numpyDizisi1 +2 , color ="blue", linewidth = 5.0, linestyle = "--", marker = "o", markersize = 10, markerfacecolor = "red" ) ## linewidth kalınlığını artırır.
```

## scatter

```python
plt.scatter(numpyDizisi1, numpyDizisi2)
```
