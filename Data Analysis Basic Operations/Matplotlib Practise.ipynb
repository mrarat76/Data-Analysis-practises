```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi=[10,20,30,30,40,50,60,72,84]
```

```python
kiloListesi =[20,60,80,85,86,87,70,76,95]
```

```python
numpyYasListesi = np.array (yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"g")
## "g" ibaresi çizginin rengini belirler.
plt.xlabel("Yas")
plt.ylabel("Kilo")
## xlabel ve ylabel metotları eksenlere isim vermek için kullanılır.
# Yukarıda ikinci kod satırında yazığımız kod yazılmasaydı plt.show() isimli  metodu kullamalıydık.
plt.title("Kilo yaş grafiği")
## title metodu grafiğe isim vermemizi sağlar.
```

```python
numpyDizisi = np.linspace(0,10,20)
```

```python
numpyDizisi
```

```python
numpyDizisi2 = numpyDizisi **3
## ** üssünü alır
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi,numpyDizisi2 ,"g-*")
## g nin yanına -- koyarsan kesikli yapar. -* koyarsan noktalı yapar.
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi,numpyDizisi2, "g-.")
plt.subplot(1,2,2)
plt.plot(numpyDizisi2, numpyDizisi,"b-*")
## (1,2,1) ve (1,2,2) den kastımız, parantezde ilk yazan sayı (1,x,x), sırayı ifade eder. İkinci yazan sayı kolonu, Üçüncü yazan sayı ise hangi grafiği çizğini ifade eder.
## Yani 1 SIRA OLACAK, 2 KOLON OLACAK, VE ŞU AN 2. GRAFİĞİ ÇİZİYORUM (1,2,2).
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.2, 0.2, 0.9 , 0.9]) ## [sol taraftaki eksenin başlangıcı , alt taraftaki eksenin başlangıç noktası, genişlik, uzunluk ]
figureAxes.plot(numpyDizisi, numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes ([0.3,0.3,0.3,0.3])

eksen1.plot(numpyDizisi, numpyDizisi2, "g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2, numpyDizisi ,"b")
eksen2.set_xlabel("X Ekseni")
eksen2.set_ylabel("Y Ekseni")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
figure3 = plt.figure()
eksen12 = figure3.add_axes([0.1,0.1,0.7,0.7])
eksen21 = figure3.add_axes ([0.7,0.1,0.2,0.5])

eksen12.plot(numpyDizisi, numpyDizisi2, "g")
eksen12.set_xlabel("X Ekseni")
eksen12.set_ylabel("Y Ekseni")
eksen12.set_title("Ana Grafik Başlık")

eksen21.plot(numpyDizisi2, numpyDizisi ,"b")
eksen21.set_xlabel("X Ekseni")
eksen21.set_ylabel("Y Ekseni")
eksen21.set_title("Küçük Grafik Başlık")
```

```python
plt.subplots() ##eksen de çıkarır.
```

```python
(benimFigure, benimEksen)= plt.subplots(nrows=1, ncols = 2) ##ekseni otomatik çizer ve figürü otomatik atar.

for eksen in benimEksen:
    eksen.plot(numpyDizisi, numpyDizisi2, "g")
    eksen.set_xlabel("X Ekseni")
plt.tight_layout()
```

```python
yeniFigur = plt.figure() ## dpi çözünürlük ve boyu, figsize, boyut. 

yeniEksen = yeniFigur.add_axes ([0.1,0.1 ,0.9,0.9])

yeniEksen.plot(numpyDizisi, numpyDizisi **2 ,label ="numpyDizisi ** 2")
yeniEksen.plot(numpyDizisi, numpyDizisi ** 3, label ="numpyDizisi ** 3" )

yeniEksen.legend()
```
